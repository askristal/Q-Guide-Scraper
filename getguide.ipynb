```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException        
from selenium.webdriver.support.select import Select
import os
import time

####################
### FILL THIS IN ###
####################

username = ''
password = ''

####################


ag = None
def scrapeIds(a,b):
    driver = webdriver.Firefox()
    driver.implicitly_wait(1)
    counter = 0
    nope = []
    for i in xrange(a,b):
        driver.get('https://webapps.fas.harvard.edu/course_evaluation_reports/fas/course_summary.html?course_id=' + str(i))
        if driver.title == 'HarvardKey Login':
            user = driver.find_element_by_name('username')
            user.send_keys(username)
            pw = driver.find_element_by_name('password')
            pw.send_keys(password)
            pw.send_keys(Keys.RETURN)

        try:
            el = driver.find_element_by_id('summaryStats') #trigger exception if not found
            element = WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.ID, "reportContent")))
            assert driver.title == 'Course Evaluations: Course Page'

            title = driver.find_element_by_tag_name('h1')
            course = title.text
            select = Select(driver.find_element_by_name('course_id'))
            selected_option = select.first_selected_option
            year = selected_option.text
            filename = year + '/' + course + str(i) + '.html'
            if not os.path.exists(os.path.dirname(filename)):
                os.makedirs(os.path.dirname(filename))
            with open(filename, "w") as f:
                f.write(driver.page_source.encode('utf8'))
            f.close()
            counter = 0
            time.sleep(0.01)
        except NoSuchElementException:
            print 'not found ' + str(i)
            nope.append(i)
            counter += 1
            if counter >= 6000:
                print counter
                print i
                print nope
                ag = nope
                raise Exception
            continue
    driver.close()
    return nope
```
